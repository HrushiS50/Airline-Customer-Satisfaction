```python
df=new_data.copy()
df
```

```python
df.head()
```

```python
df=df.drop(columns=['ID'])
```

```python
df
```

```python
df['Gender']=df['Gender'].map({'Male': 1,'Female': 0})
df['Gender']
```

```python
cols=['Customer Type','Type of Travel','Class']

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Customer Type']=le.fit_transform(df['Customer Type'])
df['Type of Travel']=le.fit_transform(df['Type of Travel'])
df['Class']=le.fit_transform(df['Class'])

```

```python
df
```

```python
df['Satisfaction']=le.fit_transform(df['Satisfaction'])
df
```

```python
data_preprocessed=df.copy()
data_preprocessed
```

```python
df=df.drop(columns='Age')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x_scaled=sc.transform(x)
x_scaled
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
log.score(x_train,y_train)
```

```python
log.score(x_test,y_test)
```

```python
y_pred=log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
Log_Reg_Train_Score=log.score(x_train,y_train)
Log_Reg_Test_Score=log.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
DTC_Classifier_Train_Score=dtc.score(x_train,y_train)
DTC_Classifier_Train_Score
```

```python
DTC_Classifier_Test_Score=dtc.score(x_test,y_test)
DTC_Classifier_Test_Score
```

```python
from sklearn.naive_bayes import GaussianNB
gn=GaussianNB()
```

```python
gn.fit(x_train,y_train)
```

```python
GN_Test_Score=gn.score(x_test,y_test)
GN_Test_Score
```

```python
GN_Train_Score=gn.score(x_train,y_train)
GN_Train_Score
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
KNN_Train_score=knn.score(x_train,y_train)
KNN_Train_score
```

```python
KNN_Test_Score=knn.score(x_test,y_test)
KNN_Test_Score
```

```python


models = ['Logistic Regression','Decision Tree', 'GausianNB', 'KNN']

train_scores = [Log_Reg_Train_Score,DTC_Classifier_Train_Score,GN_Train_Score,KNN_Train_score]
test_scores = [Log_Reg_Test_Score,DTC_Classifier_Test_Score,GN_Test_Score,KNN_Test_Score]

data = {
    'Model': models,
    'Train Score': train_scores,
    'Test Score': test_scores
}

summary_df = pd.DataFrame(data)


print(summary_df)
```
